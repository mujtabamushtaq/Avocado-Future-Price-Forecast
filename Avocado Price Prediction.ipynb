```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
from fbprophet import Prophet
```

```python
avocado_df = pd.read_csv("avocado.csv")
```

```python
avocado_df.head()
```

```python
avocado_df.tail()
```

```python
avocado_df = avocado_df.sort_values('Date')
```

```python
avocado_df
```

```python
plt.figure(figsize=(20,20)) 
plt.xticks(rotation = 90)
plt.plot(avocado_df['Date'], avocado_df['AveragePrice'])
```

```python
plt.figure(figsize=(25,12))
plt.xticks(rotation = 45)
sns.countplot(x = 'region', data= avocado_df)
```

```python
sns.countplot(x = 'year', data = avocado_df)
```

```python
avocado_prophet = avocado_df[['Date','AveragePrice']]
avocado_prophet
```

```python
avocado_prophet = avocado_prophet.rename(columns={'Date':'ds', 'AveragePrice':'y'})
```

```python
model = Prophet()
```

```python
model.fit(avocado_prophet)
```

```python
future = model.make_future_dataframe(periods = 365)
forecast = model.predict(future)
```

```python
forecast
```

```python
figure = model.plot(forecast, xlabel='Date', ylabel='Price')
```

```python
figure = model.plot_components(forecast)
```
